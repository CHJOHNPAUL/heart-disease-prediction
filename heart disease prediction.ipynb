```python
def make_prediction(Age,Sex,Chestpaintype,BP,Cholesterol,FBSover120,EKGresults,MaxHR,Exerciseangina,STdepression,SlopeofST,Numberofvesselsfluro,Thallium):
    model=pickle.load(open('mymodel','rb'))
    predictcancer=model.predict([[Age,Sex,Chestpaintype,BP,Cholesterol,FBSover120,EKGresults,MaxHR,Exerciseangina,STdepression,SlopeofST,Numberofvesselsfluro,Thallium]])
    return predictcancer

age_input = gr.Number(label = "Enter  the age ")
sex_input = gr.Number(label = "Enter  the sex")
chestpaintype_input = gr.Number(label = "Enter  the chestpaintype ")
bp_input = gr.Number(label = "Enter  the bp")
chelostral_input = gr.Number(label = "Enter  the chelostral")
fbs_input = gr.Number(label = "Enter  the fbs ")
ekg_input = gr.Number(label = "Enter  the ekg")
MaxHR_input = gr.Number(label = "Enter  the Max HR ")
Exerciseangina_input = gr.Number(label = "Enter  the Exerciseangina_input ")
STdepression_input = gr.Number(label = "Enter  the ST depression ")
SlopeofST_input = gr.Number(label = "Enter  the Slope of ST_")
Numberofvesselsfluro_input = gr.Number(label = "Enter  the Number of vessels fluro ")
Thallium_input = gr.Number(label = "Enter  the Thallium")



output = gr.Textbox()



app = gr.Interface(fn = make_prediction, inputs=[age_input,sex_input,chestpaintype_input,bp_input,chelostral_input,fbs_input,ekg_input,MaxHR_input,Exerciseangina_input,STdepression_input,SlopeofST_input,Numberofvesselsfluro_input,Thallium_input], outputs=output)
app.launch(share=True)
    
```
